```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
data = pd.read_csv('data.csv')
```

# Separate the Target from features

```python
data.head()
```

```python
cols = data.columns
cols
```

```python
y = data.diagnosis
drop_cols = ['id', 'Unnamed: 32', 'diagnosis']
x = data.drop(drop_cols, axis=1)
x.head()
```

# Plot Diagnosis Distributions

```python
ax = sns.countplot(y, label="count")
B, M = y.value_counts()
print("B: ", B)
print("M: ", M)
```

```python
x.describe()
```

# Visualizing Standized Data with Seaborn

```python
data = x
data_std = (data - data.mean()) / data.std()
data = pd.concat([y, data_std.iloc[:,0:10]], axis=1)
data = pd.melt(data, id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis',data=data, split=True,inner='quart')
plt.xticks(rotation=45)

```

# Violin Plots and Box Plots

```python
data = pd.concat([y, data_std.iloc[:,10:20]], axis=1)
data = pd.melt(data, id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis',data=data, split=True,inner='quart')
plt.xticks(rotation=45)
```

```python
data = pd.concat([y, data_std.iloc[:,20:30]], axis=1)
data = pd.melt(data, id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis',data=data, split=True,inner='quart')
plt.xticks(rotation=45)
```

#### As we can see in above plot that the features concavity_worst and concave points_worst are similar so while feature selection if they are highly correlated with each other then one of them is gonna be ignored.

```python
sns.boxplot(x='features', y='value', hue='diagnosis', data=data)
plt.xticks(rotation=45)
```

# Using Joint Plots for Feature Comparision

```python
sns.jointplot(x.loc[:, 'concavity_worst'],
            x.loc[:, 'concave points_worst'],
            kind='regg')
```

# Observing the Distribution of Values and their Variance with Swarm Plots

```python
sns.set(style='whitegrid', palette='muted')
data = x
data_std = (data - data.mean()) / data.std()
data = pd.concat([y, data_std.iloc[:,0:10]], axis=1)
data = pd.melt(data, id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features',y='value',hue='diagnosis',data=data)
plt.xticks(rotation=45)

```

```python
sns.set(style='whitegrid', palette='muted')
data = x
data_std = (data - data.mean()) / data.std()
data = pd.concat([y, data_std.iloc[:,10:20]], axis=1)
data = pd.melt(data, id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features',y='value',hue='diagnosis',data=data)
plt.xticks(rotation=45)
```
